## 4. age
Age is the only numeric feature in our dataset. It is a continious variable. <br>

<b><u>Continious Random Variable : </b></u> <br>
A continuous random variable is one which takes an infinite number of possible values. Continuous random variables are usually measurements. Examples include height, weight, the amount of sugar in an orange, the time required to run a mile.<br>

Age being a continious variable becomes one of the important feature of this dataset as we will further use wrt other variables in <b><u>Bi-variate Analysis</u></b> 



(Definition taken from Valerie J. Easton and John H. McColl's Statistics Glossary v1.1)

```python
age.describe()
```

### Statastical Measures for age 

1. The average age for the players in the years(2014,2018 and 2022 combined) is approx 26 years <br>

### Depicting the variation in the data dstribution using distribution plot:

We can see that the age distribution follows a bell shaped curve but is slightly <b>negatively skewed</b> i.e. the value of β1 > 0 (ratio of μ3 ^2 and μ2 ^3). <br>


In this kind of distribution the mean is less than the median, which is often less than the mode. We can see this clearly too as <b>Median age is 26.0</b> and <b>mean is 25.88</b>, but in this case<b> the mode and median are equal i.e. 26.</b> <br>

This distribution is also a <b>Mesokrutic</b> i.e. a distributions with zero excess kurtosis. The most prominent example of a mesokurtic distribution is the normal distribution.

```python
sns.distplot(df["age"], fit=norm)
```
